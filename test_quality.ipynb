## Observation
The metrics are proposed for a signal of a few seconds duration. If the signal is longer than that, it is recommended to analyze segments of 10 seconds.

```python
# flat signal with one minute duration with sampling rate of 1000 Hz
window_sec = 10
sampling_rate = 1000.0
signal_dummy = np.ones(int(sampling_rate*60))
window_size = int(sampling_rate*window_sec)

quality_dummy = []

for i in range(0, len(signal_dummy)-window_size, window_size):
    quality_dummy += [quality.quality_ecg(signal_dummy[i:i+window_size], 
                                          sampling_rate=sampling_rate, 
                                          methods=['Level3'])['Level3']]
print(quality_dummy)
```
